```python
 # Clustering DATASET
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([3, 1, 1, 2, 1, 6, 6, 6, 5, 6, 7, 8, 9, 8, 9, 9, 8])
x2 = np.array([5, 4, 6, 6, 5, 8, 6, 7, 6, 7, 1, 2, 1, 2, 3, 2, 3])

plt.plot()
plt.xlim([0, 10])
plt.ylim([0, 10])
plt.title('Dataset')
plt.scatter(x1, x2)
plt.show()

# Create new Plot and Data
plt.plot()
X = np.array(list(zip(x1, x2))).reshape(len(x1), 2)
colors = ['b', 'g', 'r']
markers = ['o', 'v', 's']

# KMeans Algorithm 
K = 3
kmeans_model = KMeans(n_clusters=K).fit(X)

plt.plot()
for i, l in enumerate(kmeans_model.labels_):
    plt.plot(x1[i], x2[i], color=colors[l], marker=markers[l],ls='None')
    plt.xlim([0, 10])
    plt.ylim([0, 10])

plt.show()
```

```python
# Dataset IrisPython

import pandas as pd
iris = pd.read_csv("iris.csv")
iris.head(35)
```

```python
iris.tail(15)
```

```python
X = iris.iloc[:, 0:4].values
X
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 3, init = 'random')
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
distance = kmeans.fit_transform(X)
distance
```

```python
labels = kmeans.labels_
labels
```

```python
# MÃ©todo Elbow -> O MÃ©todo Elbow, se trata de uma tÃ©cnica interessante para encontrar o valor ideal do parÃ¢metro k.
# Basicamente o que o MÃ©todo faz Ã© testar a variÃ¢ncia dos dados em relaÃ§Ã£o ao nÃºmero de Clusters.
```

```python
from sklearn.cluster import KMeans
wcss = []
i = 0 
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'random')
    kmeans.fit(X)
    print (i,kmeans.inertia_)
    wcss.append(kmeans.inertia_)  
plt.plot(range(1, 11), wcss)
plt.title('O Metodo Elbow')
plt.xlabel('Numero de Clusters')
plt.ylabel('WSS') # Within cluster sum of squares
plt.show()
```

```python
# Agrupando Novos Dados com KMeans
# Com o nosso modelo jÃ¡ criado podemos usar o Algoritmo K-means para agrupar novos dados baseado nos centroides que jÃ¡ estÃ£o definidos.
# Passando novos valores e observando os cluster que serÃ£o atribuÃ­dos.
```

```python
# Agrupando novos dados KMeansPython
data = [
        [ 4.12, 3.4, 1.6, 0.7],
        [ 5.2, 5.8, 5.2, 6.7],
        [ 3.1, 3.5, 3.3, 3.0]
    ]
kmeans.predict(data)
```

```python
# Visualizando os resultados
# Para ficar ainda mais claro Ã© interessante visualizar atravÃ©s de uma representaÃ§Ã£o grÃ¡fica os nossos dados e os centroides criados pelo K-means 
# Vejamos o cÃ³digo abaixo como Ã© simples.
```

```python
import matplotlib.pyplot as plt
 
plt.scatter(X[:, 0], X[:,1], s = 100, c = kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'red',label = 'Centroids')
plt.title('Iris Clusters and Centroids')
plt.xlabel('SepalLength')
plt.ylabel('SepalWidth')
plt.legend()
 
plt.show()
```
